```python
import numpy as np
import pandas as pd
import yfinance as yf
from ta.trend import ADXIndicator
from dataclasses import dataclass, field
from typing import List, Optional, Set, Dict
import tensorflow as tf
from tensorflow.keras import layers, Model
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.feature_selection import mutual_info_regression
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import seaborn as sns
import os
import logging
from itertools import combinations

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

@dataclass
class DataConfig:
    tickers: List[str] = field(default_factory=lambda: (
        "BAJAJ-AUTO.NS BRITANNIA.NS HINDUNILVR.NS INFY.NS ITC.NS "
        "M&M.NS RELIANCE.NS SBIN.NS TCS.NS WIPRO.NS"
    ).split())
    start: str = "2005-01-01"
    end: Optional[str] = "2025-01-01"
    interval: str = "1d"
    use_adjusted: bool = True

@dataclass
class TaskConfig:
    target: str = "log_ret"
    horizon: int = 1
    seq_len: int = 5
    # ENHANCED FEATURE SET
    feature_cols: List[str] = field(default_factory=lambda: [
        "log_ret", 
        "log_ret_5d",
        "log_ret_8d",
        "log_ret_10d",
        # "std_ret",
        # "adx10",
        # "mr_zscore",
        # "vol_ratio",
        # "price_range",
        "ema10_ratio",
        "ema20_ratio", 
        # "ema128_ratio", 
        # "ema200_ratio",
        # "roc10", 
        "rsi14",
        "macd_over_close", 
        # "atr14_over_close",
        "vol20",
        # "vol60",
        "force1",
        # "dpo20",
        "stoch_k", 
        "williams_r", 
        "mfi"
    ])

@dataclass
class TrainConfig:
    train_ratio: float = 0.70
    val_ratio: float = 0.15
    seed: int = 237
    batch_size: int = 64
    epochs: int = 100
    model_name: str = "./Stock-project/Final/enhanced_stockformer_best.keras"
    normalization_window: int = 5

@dataclass
class ModelConfig:
    embed_dim: int = 128
    num_heads: int = 4
    ff_dim: int = 64
    num_layers: int = 3
    dropout_embed: float = 0.1
    dropout_attn: float = 0.1
    dropout_ffn: float = 0.1
    dropout_head: float = 0.2
    lr: float = 1e-4
    optimizer: str = "adamw"
    loss: str = "huber"
```

```python
def load_data(cfg: DataConfig) -> pd.DataFrame:
    logger.info(f"Loading data for {len(cfg.tickers)} tickers from {cfg.start} to {cfg.end}")
    
    data = yf.download(
        tickers=cfg.tickers,
        start=cfg.start,
        end=cfg.end,
        interval=cfg.interval,
        auto_adjust=cfg.use_adjusted,
        group_by="ticker",
        progress=False
    )
    
    if not isinstance(data.columns, pd.MultiIndex):
        ticker = cfg.tickers[0] if len(cfg.tickers)==1 else "UNKNOWN"
        df = data.copy()
        df["Ticker"] = ticker
        df = df.reset_index().set_index(["Date","Ticker"]).sort_index()
        logger.info(f"Single ticker data loaded: {df.shape}")
        return df
    
    cols = data.columns
    lvl0 = set(cols.get_level_values(0))
    lvl1 = set(cols.get_level_values(1))
    tickers_set = set(cfg.tickers)
    
    if tickers_set & lvl0:
        stack_level = 0
    elif tickers_set & lvl1:
        stack_level = 1
    else:
        attr_candidates = {"Open","High","Low","Close","Adj Close","Volume"}
        stack_level = 1 if (attr_candidates & lvl0) else 0
    
    stacked = data.stack(level=stack_level).reset_index()
    
    ticker_col = None
    for c in stacked.columns[:3]:
        if set(stacked[c].astype(str).unique()) & tickers_set:
            ticker_col = c
            break
    if ticker_col is None:
        ticker_col = stacked.columns[1]
    
    stacked = stacked.rename(columns={ticker_col: "Ticker"})
    stacked = stacked.set_index(["Date","Ticker"]).sort_index()
    
    logger.info(f"Multi-ticker data loaded: {stacked.shape}")
    logger.info(f"Date range: {stacked.index.get_level_values(0).min()} to {stacked.index.get_level_values(0).max()}")
    
    return stacked

def add_enhanced_features(df: pd.DataFrame) -> pd.DataFrame:
    logger.info("Starting enhanced feature engineering...")
    
    def _feat(g: pd.DataFrame) -> pd.DataFrame:
        g = g.sort_index()
        C = g["Close"]
        H = g["High"]
        L = g["Low"]
        V = g["Volume"]
        O = g["Open"]
                
        # Core price features 
        g["log_ret"] = np.log(C.shift(1) / C.shift(2))  # Previous day return
        g["price_range"] = ((H - L) / C).shift(1)  # PREVIOUS day's range

        # Multi-Horizon Log Returns
        for h in [5, 8, 10]:
            g[f"log_ret_{h}d"] = np.log(C.shift(1) / C.shift(1 + h)).fillna(0)

        # Risk-Adjusted Return 
        g["vol20"] = np.log(C.shift(1) / C.shift(2)).rolling(20).std().shift(1)
        g["std_ret"] = (np.log(C.shift(1) / C.shift(2)) / (g["vol20"] + 1e-9)).fillna(0)

        # ADX(14)
        adx_ind = ADXIndicator(high=H, low=L, close=C, window=10, fillna=False)
        g["adx10"] = adx_ind.adx().shift(1).fillna(0)

        # Volatility Ratio
        vol60 = np.log(C.shift(1) / C.shift(2)).rolling(60).std().shift(1)
        g["vol_ratio"] = (g["vol20"] / (vol60 + 1e-9)).fillna(0)

        # Mean-Reversion Z-Score
        rolling_mean20 = C.shift(1).rolling(20).mean()
        rolling_std20 = C.shift(1).rolling(20).std()
        g["mr_zscore"] = ((C.shift(1) - rolling_mean20) / (rolling_std20 + 1e-9)).fillna(0)
        
        # Enhanced moving averages 
        for span in [10, 20, 128, 200]:
            # Calculate on past prices (excluding current day)
            ema = C.shift(1).ewm(span=span, adjust=False).mean()
            g[f"ema{span}_ratio"] = C.shift(1) / (ema + 1e-9)
        
        # Momentum indicators 
        g["roc10"] = C.shift(1).pct_change(10)
        
        # RSI 
        past_returns = np.log(C.shift(1) / C.shift(2))
        delta = past_returns.diff()
        up = delta.clip(lower=0)
        down = -delta.clip(upper=0)
        roll_up = up.ewm(span=14, adjust=False).mean()
        roll_down = down.ewm(span=14, adjust=False).mean()
        rs = roll_up / (roll_down + 1e-9)
        g["rsi14"] = 100 - (100 / (1 + rs))
        
        # MACD 
        past_close = C.shift(1)
        ema14 = past_close.ewm(span=14, adjust=False).mean()
        ema26 = past_close.ewm(span=26, adjust=False).mean()
        macd = ema14 - ema26
        g["macd_over_close"] = macd / (past_close + 1e-9)
        
        # ATR 
        H_lag = H.shift(1)
        L_lag = L.shift(1) 
        C_lag = C.shift(1)
        C_lag2 = C.shift(2)
        
        high_low = H_lag - L_lag
        high_close = (H_lag - C_lag2).abs()
        low_close = (L_lag - C_lag2).abs()
        tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        atr14 = tr.ewm(span=14, adjust=False).mean()
        g["atr14_over_close"] = atr14 / (C_lag + 1e-9)
        
        # Volume indicators 
        g["force1"] = ((C.shift(1) - C.shift(2)) * V.shift(1))
        
        # DPO (using past data)
        sma14_past = C.shift(1).rolling(window=20).mean()
        g["dpo20"] = C.shift(10) - sma14_past  # Shift by half the period
        
        # Stochastic 
        H_past = H.shift(1)
        L_past = L.shift(1)
        C_past = C.shift(1)
        low_min = L_past.rolling(window=14).min()
        high_max = H_past.rolling(window=14).max()
        g["stoch_k"] = 100 * (C_past - low_min) / (high_max - low_min + 1e-9)
        
        # Williams %R 
        low_min = L_past.rolling(window=14).min()
        high_max = H_past.rolling(window=14).max()
        g["williams_r"] = -100 * (high_max - C_past) / (high_max - low_min + 1e-9)
        
        # MFI 
        typical_price_past = ((H + L + C) / 3).shift(1)
        money_flow = typical_price_past * V.shift(1)
        
        price_change = typical_price_past.diff()
        positive_flow = money_flow.where(price_change > 0, 0)
        negative_flow = money_flow.where(price_change < 0, 0)
        
        positive_mf = positive_flow.rolling(window=14).sum()
        negative_mf = negative_flow.rolling(window=14).sum()
        mf_ratio = positive_mf / (negative_mf + 1e-9)
        g["mfi"] = 100 - (100 / (1 + mf_ratio))
        
        return g
    
    df_feats = df.groupby(level="Ticker", group_keys=False).apply(_feat)
    df_feats = df_feats.dropna()
    
    logger.info(f"Feature engineering completed. Shape: {df_feats.shape}")
    logger.info(f"Features created: {[col for col in df_feats.columns if col not in ['Open', 'High', 'Low', 'Close', 'Volume']]}")
    
    return df_feats

def add_targets(df: pd.DataFrame, task_cfg) -> pd.DataFrame:
    logger.info("Adding target variables...")
    
    tgt = task_cfg.target  
    h = task_cfg.horizon   
    
    df = df.copy()
    df["target"] = df.groupby(level="Ticker")[tgt].shift(-h)
    df = df.dropna(subset=["target"])
    
    logger.info(f"Targets added. Final shape: {df.shape}")
    
    return df
```

```python
class PerTickerRollingZNormalizer:
    def __init__(self, feature_cols: List[str], target_col: str = "target", window: int = 60):
        self.feature_cols = feature_cols
        self.target_col = target_col
        self.window = window
        self.last_windows = {}
        logger.info(f"Initialized per-ticker Rolling-Z normalizer with window={window}")

    def fit_transform(self, df: pd.DataFrame) -> pd.DataFrame:
        logger.info("Starting per-ticker rolling z-score normalization...")
        normalized = []

        for ticker, group in df.groupby(level="Ticker"):
            logger.info(f"   Normalizing {ticker}...")
            g = group.sort_index().copy()

            # Rolling z-score for each feature
            for col in self.feature_cols:
                roll_mean = g[col].rolling(window=self.window, min_periods=1).mean()
                roll_std  = g[col].rolling(window=self.window, min_periods=1).std().add(1e-8)
                g[col] = (g[col] - roll_mean) / roll_std

            tm = g[self.target_col].rolling(window=self.window, min_periods=1).mean()
            ts = g[self.target_col].rolling(window=self.window, min_periods=1).std().add(1e-8)
            g[self.target_col] = (g[self.target_col] - tm) / ts

            self.last_windows[ticker] = {
                **{col: (roll_mean.iloc[-1], roll_std.iloc[-1]) for col in self.feature_cols},
                self.target_col: (tm.iloc[-1], ts.iloc[-1])
            }

            normalized.append(g.dropna())

        result = pd.concat(normalized).sort_index()
        logger.info(f"Rolling-Z normalization completed. Shape: {result.shape}")
        return result

    def inverse_transform_target(self, ticker: str, y_scaled: np.ndarray) -> np.ndarray:
        mean_, std_ = self.last_windows[ticker][self.target_col]
        return y_scaled * std_ + mean_

    
    def _rolling_zscore(self, series: pd.Series) -> pd.Series:
        rolling_mean = series.rolling(window=self.window, min_periods=self.window//2).mean()
        rolling_std = series.rolling(window=self.window, min_periods=self.window//2).std()
        
        return (series - rolling_mean) / (rolling_std + 1e-8)
```

```python
def make_splits_and_datasets(df: pd.DataFrame, task_cfg: TaskConfig, train_cfg: TrainConfig):
    logger.info("Creating data splits and datasets...")
    
    # split per-ticker chronologically
    dfs_train, dfs_val, dfs_test = [], [], []
    for ticker, g in df.groupby(level="Ticker"):
        g = g.sort_index()
        n = len(g)
        n_train = int(n * train_cfg.train_ratio)
        n_val = int(n * train_cfg.val_ratio)
        g_train = g.iloc[:n_train]
        g_val = g.iloc[n_train:n_train+n_val]
        g_test = g.iloc[n_train+n_val:]
        dfs_train.append(g_train); dfs_val.append(g_val); dfs_test.append(g_test)
        
        logger.info(f"   {ticker}: Train={len(g_train)}, Val={len(g_val)}, Test={len(g_test)}")
    
    df_train = pd.concat(dfs_train).sort_index()
    df_val = pd.concat(dfs_val).sort_index()
    df_test = pd.concat(dfs_test).sort_index()
    
    logger.info(f"Split sizes - Train: {len(df_train)}, Val: {len(df_val)}, Test: {len(df_test)}")
    
    # normalizer (fit on train)
    normalizer = PerTickerRollingZNormalizer(
        task_cfg.feature_cols, 
        target_col="target", 
        window=train_cfg.normalization_window
    )
    
    df_train_s = normalizer.fit_transform(df_train)
    df_val_s = normalizer.fit_transform(df_val)
    df_test_s = normalizer.fit_transform(df_test)
    
    def build_sequences(df_s):
        Xs, ys = [], []
        meta = []  # optional metadata: (ticker, date_index_of_target)
        for ticker, g in df_s.groupby(level="Ticker"):
            g = g.sort_index()
            X = g[task_cfg.feature_cols].values.astype(np.float32)
            y = g["target"].values.astype(np.float32)
            L = len(g)
            seq = task_cfg.seq_len
            h = task_cfg.horizon
            for i in range(0, L - seq - h + 1):
                Xs.append(X[i:i+seq])
                ys.append(y[i+seq + h - 1])
                meta.append((ticker, g.index[i+seq + h - 1][0]))  # (Ticker, Date)
        if len(Xs)==0:
            return np.empty((0, task_cfg.seq_len, len(task_cfg.feature_cols)), dtype=np.float32), np.empty((0,), dtype=np.float32), []
        return np.stack(Xs), np.stack(ys).reshape(-1), meta
    
    X_train, y_train, meta_train = build_sequences(df_train_s)
    X_val,   y_val,   meta_val   = build_sequences(df_val_s)
    X_test,  y_test,  meta_test  = build_sequences(df_test_s)
    
    logger.info(f"Sequence shapes - Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")
    
    def to_tfds(X, y, shuffle=False):
        ds = tf.data.Dataset.from_tensor_slices((X, y))
        if shuffle:
            ds = ds.shuffle(1000, seed=train_cfg.seed)
        ds = ds.batch(train_cfg.batch_size).prefetch(tf.data.AUTOTUNE)
        return ds
    
    ds_train = to_tfds(X_train, y_train, shuffle=True)
    ds_val   = to_tfds(X_val, y_val, shuffle=False)
    ds_test  = to_tfds(X_test, y_test, shuffle=False)
    
    splits = {
        "df_train": df_train, "df_val": df_val, "df_test": df_test,
        "meta_train": meta_train, "meta_val": meta_val, "meta_test": meta_test,
        "X_train": X_train, "y_train": y_train,
        "X_val": X_val, "y_val": y_val,
        "X_test": X_test, "y_test": y_test
    }
    return ds_train, ds_val, ds_test, normalizer, splits
```

```python
class LearnablePositionalEncoding(layers.Layer):
    def __init__(self, maxlen: int, embed_dim: int, **kwargs):
        super().__init__(**kwargs)
        self.pos_emb = layers.Embedding(input_dim=maxlen, output_dim=embed_dim)
    
    def call(self, x):
        seq_len = tf.shape(x)[1]
        positions = tf.range(seq_len)
        pos = self.pos_emb(positions)
        return x + pos

class CLSToken(tf.keras.layers.Layer):
    def __init__(self, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.embed_dim = embed_dim
    
    def build(self, input_shape):
        self.cls = self.add_weight(
            name="cls_token",                  
            shape=(1, 1, self.embed_dim),
            initializer="zeros",
            trainable=True
        )
    
    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        cls_tokens = tf.tile(self.cls, [batch_size, 1, 1])
        return tf.concat([cls_tokens, inputs], axis=1)

class CLSPool(layers.Layer):
    def call(self, x):
        return x[:,0,:]

class TransformerEncoderBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout_attn=0.1, dropout_ffn=0.1, **kwargs):
        super().__init__(**kwargs)
        assert embed_dim % num_heads == 0
        self.norm1 = layers.LayerNormalization(epsilon=1e-6)
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim//num_heads, dropout=dropout_attn)
        self.drop1 = layers.Dropout(dropout_attn)
        self.norm2 = layers.LayerNormalization(epsilon=1e-6)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim, activation=tf.keras.activations.gelu),
            layers.Dense(embed_dim),
        ])
        self.drop2 = layers.Dropout(dropout_ffn)
    
    def call(self, x, training=False):
        h = self.norm1(x)
        a = self.att(h, h, training=training)
        a = self.drop1(a, training=training)
        x = x + a
        h2 = self.norm2(x)
        f = self.ffn(h2, training=training)
        f = self.drop2(f, training=training)
        return x + f
```

```python
class DirectionalAccuracy(tf.keras.metrics.Metric):
    def __init__(self, name="directional_accuracy", **kwargs):
        super().__init__(name=name, **kwargs)
        self.correct = self.add_weight(name="correct", shape=(), initializer="zeros")
        self.total = self.add_weight(name="total", shape=(), initializer="zeros")
    
    def update_state(self, y_true, y_pred, sample_weight=None):
        y_true = tf.cast(y_true, self.dtype)
        y_pred = tf.cast(y_pred, self.dtype)
        true_dir = tf.sign(y_true)
        pred_dir = tf.sign(y_pred)
        correct = tf.reduce_sum(tf.cast(tf.equal(true_dir, pred_dir), self.dtype))
        total = tf.cast(tf.size(y_true), self.dtype)
        self.correct.assign_add(correct)
        self.total.assign_add(total)
    
    def result(self):
        return self.correct / (self.total + tf.keras.backend.epsilon())
    
    def reset_state(self):
        self.correct.assign(0)
        self.total.assign(0)

class RSquare(tf.keras.metrics.Metric):
    def __init__(self, name="r2_score", **kwargs):
        super().__init__(name=name, **kwargs)
        self.sse = self.add_weight(name="sse", shape=(), initializer="zeros")
        self.sst = self.add_weight(name="sst", shape=(), initializer="zeros")
    
    def update_state(self, y_true, y_pred, sample_weight=None):
        y_true = tf.cast(y_true, self.dtype)
        y_pred = tf.cast(y_pred, self.dtype)
        ss_res = tf.reduce_sum(tf.square(y_true - y_pred))
        ss_tot = tf.reduce_sum(tf.square(y_true - tf.reduce_mean(y_true)))
        self.sse.assign_add(ss_res)
        self.sst.assign_add(ss_tot)
    
    def result(self):
        return 1 - (self.sse / (self.sst + tf.keras.backend.epsilon()))
    
    def reset_state(self):
        self.sse.assign(0)
        self.sst.assign(0)


def build_enhanced_transformer(seq_len, n_features, model_cfg):
    logger.info("Building enhanced transformer model...")
    
    inp = layers.Input(shape=(seq_len, n_features), name="inputs")
    
    # Standard projection
    x = layers.Dense(model_cfg.embed_dim, name="proj")(inp)
    pe_input_length = seq_len
    
    # CLS token insertion
    x = CLSToken(model_cfg.embed_dim, name="cls_token")(x)
    pe_input_length += 1
    
    # Learnable positional encoding with fixed max length
    x = LearnablePositionalEncoding(pe_input_length, model_cfg.embed_dim, name="pos_enc")(x)
    x = layers.Dropout(model_cfg.dropout_embed)(x)
    
    # Encoder blocks
    for i in range(model_cfg.num_layers):
        x = TransformerEncoderBlock(
            embed_dim=model_cfg.embed_dim,
            num_heads=model_cfg.num_heads,
            ff_dim=model_cfg.ff_dim,
            dropout_attn=model_cfg.dropout_attn,
            dropout_ffn=model_cfg.dropout_ffn,
            name=f"encoder_{i+1}"
        )(x)
    
    # Pooling
    x = CLSPool(name="cls_pool")(x)
    
    x = layers.Dropout(model_cfg.dropout_head)(x)
    out = layers.Dense(1, name="head")(x)
    
    model = Model(inp, out, name="EnhancedStockTransformer")
    
    if model_cfg.optimizer == "adamw":
        opt = tf.keras.optimizers.AdamW(learning_rate=model_cfg.lr)
    else:
        opt = tf.keras.optimizers.Adam(learning_rate=model_cfg.lr)
    
    metrics_list = [
        tf.keras.metrics.MeanAbsoluteError(name="mae"),
        tf.keras.metrics.RootMeanSquaredError(name="rmse"),
        tf.keras.metrics.MeanAbsolutePercentageError(name="mape"),
        DirectionalAccuracy(),
        RSquare()
    ]
    
    if model_cfg.loss == "huber":
        loss_fn = tf.keras.losses.Huber()
    elif model_cfg.loss == "mse":
        loss_fn = tf.keras.losses.MeanSquaredError()
    else:
        loss_fn = tf.keras.losses.MeanAbsoluteError()
    
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics_list)
    
    logger.info(f"Model built with {model.count_params():,} parameters")
    
    return model
```

```python
def calculate_permutation_importance(model, X_test, y_test, feature_names, n_repeats=10):
    logger.info("Calculating permutation importance...")
    
    def model_predict(X):
        return model.predict(X, verbose=0).flatten()
    
    #calculate baseline score
    baseline_score = model.evaluate(X_test, y_test, verbose=0)[0]  # Loss
    
    # Calculate permutation importance
    importances = []
    
    for i, feature_name in enumerate(feature_names):
        logger.info(f"   Processing feature: {feature_name}")
        
        scores = []
        for repeat in range(n_repeats):
            X_permuted = X_test.copy()
            np.random.shuffle(X_permuted[:, :, i])
            
            permuted_score = model.evaluate(X_permuted, y_test, verbose=0)[0]
            scores.append(permuted_score - baseline_score)
        
        importance = np.mean(scores)
        std = np.std(scores)
        importances.append({'feature': feature_name, 'importance': importance, 'std': std})
    
    importance_df = pd.DataFrame(importances).sort_values('importance', ascending=False)
    
    logger.info("Permutation Importance Results:")
    for _, row in importance_df.iterrows():
        logger.info(f"   {row['feature']:20s}: {row['importance']:8.6f} ± {row['std']:6.4f}")
    
    return importance_df
```

```python
def calculate_enhanced_metrics(model, ds_test, meta_test=None):
    logger.info("Calculating enhanced metrics...")
    
    y_true_list, y_pred_list = [], []
    
    for X_batch, y_batch in ds_test:
        preds = model.predict(X_batch, verbose=0)
        y_true_list.append(y_batch.numpy())
        y_pred_list.append(preds)
    
    if len(y_true_list) == 0:
        return {}
    
    y_true = np.concatenate(y_true_list).flatten()
    y_pred = np.concatenate(y_pred_list).flatten()
    
    # Standard regression metrics
    mae = float(np.mean(np.abs(y_true - y_pred)))
    rmse = float(np.sqrt(np.mean((y_true - y_pred)**2)))
    mape = float(np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100)
    
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    r2 = float(1 - (ss_res / (ss_tot + 1e-8)))
    
    # FINANCIAL-SPECIFIC METRICS
    
    true_dir = np.sign(y_true)
    pred_dir = np.sign(y_pred)
    directional_accuracy = float(np.mean(true_dir == pred_dir))
    
    smape = float(200 * np.mean(np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true) + 1e-8)))
    
    strategy_returns = pred_dir * y_true  
    avg_return = float(np.mean(strategy_returns))
    volatility = float(np.std(strategy_returns))
    sharpe_ratio = float(avg_return / (volatility + 1e-8))
    
    profitable_trades = np.sum(strategy_returns > 0)
    total_trades = len(strategy_returns)
    hit_rate = float(profitable_trades / total_trades)
    
    cumulative_returns = np.cumprod(1 + strategy_returns) - 1
    peak = np.maximum.accumulate(cumulative_returns)
    drawdown = (cumulative_returns - peak) / (1 + peak)
    max_drawdown = float(np.min(drawdown))
    
    results = {
        "MAE": mae,
        "RMSE": rmse, 
        "MAPE": mape,
        "SMAPE": smape,
        "R_squared": r2,
        "Directional_Accuracy": directional_accuracy,
        "Hit_Rate": hit_rate,
        "Avg_Return": avg_return,
        "Volatility": volatility,
        "Sharpe_Ratio": sharpe_ratio,
        "Max_Drawdown": max_drawdown,
        "Total_Return": float(cumulative_returns[-1])
    }
    
    return results
```
